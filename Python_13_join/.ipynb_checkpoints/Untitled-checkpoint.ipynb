**Задание 1**1. Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
df = pd.read_csv('ratings.csv')
visits_grouped = df.groupby('userId').agg({'timestamp':['min', 'max', 'count']}).reset_index()

visits_grouped_filtred = visits_grouped[visits_grouped['timestamp']['count'] > 100]
visits_grouped_filtred ['life'] = visits_grouped['timestamp']['max'] - visits_grouped['timestamp']['min']
visits_grouped_filtred.life.mean()

#df
```
