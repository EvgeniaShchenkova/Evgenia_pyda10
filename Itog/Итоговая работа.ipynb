**10.Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию.**

Гипотеза: Уровень удовлетворенности сотрудника зависит от среднего колличества часов в месяц

Доверительный интервал:

```python
st_deviation = statistics.pstdev(df.satisfaction_level)
mean_val = statistics.mean(df.satisfaction_level)
```

```python
len(df.satisfaction_level[(df.satisfaction_level>=mean_val-2*st_deviation) & (df.satisfaction_level<=mean_val + 2*st_deviation)])/len(df.satisfaction_level)
```

94% выборки находится в пределах доверительного интервала.

Ранговая корреляция Спирмена

Проверяет, имеют ли два образца монотонные отношения.

H0: Зависимость между зарплатой и удовлетворенностью есть 
H1: Никакой зависимости нет

```python
# Example of the Spearman's Rank Correlation Test
from scipy.stats import spearmanr

stat, p = spearmanr(df.satisfaction_level, df.average_montly_hours)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

Таким образом, мы можем принять нулевую гипотезу о том, что есть зависимость между зарплатой сотрудника и его удовлетворенностью и эта зависимость положительна (т.е. чем больше зп - тем больше удовлетворенность

Тогда предположим, что уровень зарплаты влияет на увольнение сотрудника
H0: Зависимость между зарплатой и увольнением есть
H1: Никакой зависимости нет

```python
# Example of the Spearman's Rank Correlation Test
from scipy.stats import spearmanr

stat, p = spearmanr(df.average_montly_hours, df.left)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

Таким образом, мы можем принять нулевую гипотезу о том, что есть зависимость между зарплатой сотрудника и тем, что он покинул компанию
