```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
```

Для построения графиков были выбраны результаты опробования скважин, с содержаниями 3-х элементов: Ag, Pb, Zn.
в начале можем посмотреть отношение между всеми тремя элементами и их распределения. Для этого построим график типа pairplot.


```python
assay = pd.read_excel('HW.xls')
sns_plot = sns.pairplot(
    assay[['Ag', 'Pb', 'Zn']])
sns_plot
```

Затем можно построить точечную диаграмму только с двумя элементамии по осям Х и У, для того, чтобы посмотреть как они коррелируют друг с другом.

```python
assay = pd.read_excel('HW.xls')
assay 
assay.plot.scatter(x='Ag', y='Pb', s=30,
                      cmap='viridis')
```

Такое облако точек говорит о достаточно неплохой корреляции серебра и свинца лруг с другом.

Построим гистрограмму для длин интервалов. Это поможет нам определить к какой длине композитных интервалов привести наше опробование для проведения геостатистического анализа.

```python
assay = pd.read_excel('HW.xls')
assay
ax = assay['Длина РИ'].hist(bins=30)
```

Из графика видно, что наиболее распространенная длина интервала - 0.5

Для того, чтобы оценить распределние содержаний серебра в пробах отлично подойдет гистрограмма. 

```python
sns.distplot(assay.Ag)
```

На получившемся графике отлично видно, что наибольшее кол-во проб с очень маленькими содержаниями и имеет логнормальное распределние.
